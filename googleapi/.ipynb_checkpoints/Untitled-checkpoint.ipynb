```python
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys

url = []
for i in gu:
    browser = webdriver.Chrome("/Users/yhj/Desktop/study/scrapper/chromedriver")
    browser.get("https://www.google.com/maps/")
    elem = browser.find_element_by_id('searchboxinput')
    elem.clear()
    elem.send_keys(i)
    elem = browser.find_element_by_id('searchbox-searchbutton')
    elem.click()
    time.sleep(3)
    current_url = browser.current_url
    url.append(current_url)
    browser.close()
```

```python
url_test = []
print(gu[182])
p = re.compile("@.......................")
for i in range(len(url)):
    test = p.search(url[i])
    if test:
        url_test.append(test.group().replace("@", ""))
    else:
        url_test.append("35.3398712,126.6150124,")
        print(i)
for j in range(len(url_test)):
    cnt = 0
    for k in range(len(url_test[j])):
        if url_test[j][k] == ",":
            cnt += 1
            if cnt == 2:
                break
    url_test[j] = url_test[j][:k]
for p in url_test:
    print(p)
        
```

```python
gu_location = pd.DataFrame(
    {'행정구역' : gu,
     'test': url_test})
gu_location['위도'] = gu_location.test.str.split(',').str[0]
gu_location['경도'] = gu_location.test.str.split(',').str[1]
```

```python
gu_gps = gu_location[['행정구역', '위도', '경도']]
```
